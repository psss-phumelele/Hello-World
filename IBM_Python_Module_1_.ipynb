```python
#creating headers for our data
headers= ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style","drive-wheels",
          "engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
          "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio",
          "horsepower","peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n",headers)

```

```python
df.columns=headers
df.columns
```

```python
df.head(10)  #view now includes the headers
```

```python
#our table has ? - we will replace them with NaN df1 - is now our new table
df1=df.replace('?',np.nan)
```

```python
#we wil drop missing values along the price column
df = df1.dropna(subset=["price"],axis=0)
df1.head(20)
```
